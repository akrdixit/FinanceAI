```python
from gdeltdoc import GdeltDoc, Filters
import pandas as pd
import json
from datetime import datetime, timedelta
```

```python
df_nifty=pd.read_csv('data/ind_nifty50list.csv')
```

```python
df_nifty.columns
```

```python
df_nifty[['Company Name', 'Symbol']]
```

```python
def gdelt_data(keyword, start_date, end_date):
        f = Filters(
        keyword = keyword,
        start_date = start_date,
        end_date = end_date,
        domain_exact= ['moneycontrol.com', 'economictimes.indiatimes.com']
        )

        gd = GdeltDoc()

        # Search for articles matching the filters
        articles = gd.article_search(f)

        return articles 
```

```python
df=[]
for company in df_nifty['Company Name']:

    end_date="2024-04-28"
    original_end_date=end_date
    keyword= company
    sdate_error=[]
    edate_error=[]
    company_error=[]

    for j in range(7):

        try:


            date_obj = datetime.strptime(end_date, "%Y-%m-%d")
            start_date = date_obj - timedelta(days=1)
            start_date=start_date.strftime("%Y-%m-%d")
            # print('start_date', start_date)

            articles=gdelt_data(keyword, start_date, end_date)
            # print('************')
            articles['start_date'] = start_date
            articles['end_date'] = end_date
            articles['company'] = keyword
            print([('start_date', start_date), ('end_date', end_date), ('company', keyword)])
            df.append(articles)
            end_date = start_date
        except ValueError:

            sdate_error.append(start_date)
            edate_error.append(end_date)
            company_error.append(keyword)
            print('ohh error')
            end_date = start_date
            date_obj = datetime.strptime(end_date, "%Y-%m-%d")
            start_date = date_obj - timedelta(days=1)
            start_date=start_date.strftime("%Y-%m-%d")
    
```

```python
df1=pd.concat(df)
```

```python
df1.to_csv('data/aweekstocks.csv', index=False)
```
